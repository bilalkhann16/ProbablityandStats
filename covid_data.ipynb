```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(confirmed)
plt.show()
```

```python
plt.figure(figsize=(10, 5)) 
plt.plot(confirmed)
plt.title("Covid Cases Visualization")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15, 5)) 
x=data['Country/Region'].head(10)
y=data['Confirmed'].head(10)
plt.plot(x,y)
plt.title("Covid Cases Visualization")
plt.xlabel("Countires")
plt.ylabel("Total Confired Cases")
plt.grid()
plt.show()
```

```python
#Plotting the countries with most active cases.
s=data[['Country/Region','Confirmed']].sort_values('Confirmed',ascending=False)
s=s.head(11)
print (s)
```

```python
plt.figure(figsize=(15,5))
x=s['Country/Region']
y=s['Confirmed']
plt.plot(x,y)

plt.xlabel("Country/Region")
plt.ylabel("Total confirmed cases")
plt.title("Top 10 Countires With Most Active Cases WorldWide")
plt.grid()
plt.show()
```
